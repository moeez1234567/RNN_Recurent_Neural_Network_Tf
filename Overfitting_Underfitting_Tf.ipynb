```python
import tensorflow as tf
import matplotlib.pyplot as plt 

```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
def simple_built():
    input_layer = tf.keras.layers.Input(shape = (28,28,1))
    model = tf.keras.models.Sequential([
        input_layer,
        tf.keras.layers.Conv2D(filters = 16, kernel_size = (3,3), activation = 'relu'),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(10, activation='softmax')
    ])

    model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
    return model
```

```python
simple_model = simple_built()

histroy = simple_model.fit(X_train, y_train, epochs = 10, verbose = 1, validation_split=0.2)
```

```python
# Final Model

X_train = X_train / 255.0
X_test = X_test /255.0
```

```python
# Adiing regulizer which can work same as dropout layer but another way 
#regulizer.l1 can finih some weights exactly zero
#regulizer.l2 can shrink some weight that it cant overfitt
#reglarizer use both l1,l2 shriks some weights and finish 
from tensorflow.keras import regularizers
```

```python
# Reduce lr
from tensorflow.keras.optimizers import Adam

optimizer = Adam(learning_rate = 0.0001)
```

```python
def final_model():
    input_layer = tf.keras.layers.Input(shape = (28,28,1))
    model = tf.keras.models.Sequential([
    input_layer,
    tf.keras.layers.Conv2D(filters = 32, kernel_size = (3,3), use_bias = False, kernel_regularizer= regularizers.l2(0.01)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation('relu'),

    tf.keras.layers.Conv2D(filters=64, kernel_size = (3,3), use_bias=False, kernel_regularizer = regularizers.l1(0.001)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation('relu'),

    tf.keras.layers.Conv2D(filters=128, kernel_size = (3,3), use_bias=False, kernel_regularizer = regularizers.l1(0.001)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation('relu'),
  

    tf.keras.layers.GlobalAveragePooling2D(),
    
    tf.keras.layers.Dense(64, kernel_regularizer = regularizers.l1_l2(l1 = 0.001, l2 = 0.001)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation('relu'),

    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(10, activation='softmax')

])

    model.compile(optimizer = optimizer, loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
    return model
    
    
```

```python
f = final_model()
```

```python
histroy = f.fit(X_train, y_train, epochs = 10, verbose = 1,batch_size = 64, validation_split=0.2 )
```
